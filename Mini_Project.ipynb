```
# Library Imports

import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize
import pandas as pd
```

### ***DATA GATHERING*** 



```
cd /content/drive/MyDrive/Images
```

```
#Importing Data from Drive 
DATA = '/content/drive/MyDrive/Images'
```

### **DATA PREPROCESSING**

```
Category = ['dog','fish','butterfly']
target = []
images = []
flat_data =[]
for i in Category:

  #Label Encoding for Category of Images
  class_num = Category.index(i)
  path = os.path.join(DATA,i)
  print(path)
  for img in os.listdir(path):
    img_array = imread(os.path.join(path,img))
    
    #Data Normalization
    # 0-255 will be normalized 0-1
    img_resized = resize(img_array,(150,150,3)) 

    #Flattening of Data 
    flat_data.append(img_resized.flatten())
    images.append(img_resized)
    target.append(class_num)

flat_data  = np.array(flat_data)
target = np.array(target)
images = np.array(images)
```

```
#Data Frame Creation

df = pd.DataFrame(flat_data)
df
```

```
#Adding Target as our new column

df['Target'] = target
df
```

### **APPLYING SVM ALGORITHM**

```
#Splitting the DataFrame into Testing And Training

x = df.iloc[:,:67500]
y = df.iloc[:,67500]
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
```

```
#Importing libraries for SVM Algorithm and Applying GridSearch CV

from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

svm_classifier = SVC()

parameters = [{'C':[1,10,100,1000],'kernel': ['linear'],
               'C':[1,10,100,1000],'kernel': ['rbf'],'gamma':[0.0,0.001,0.1,10,20,100],
               'C':[1,10,100,1000],'kernel': ['rbf'],'gamma':['scale']}]
grid_search = GridSearchCV(estimator =svm_classifier,param_grid = parameters,scoring='accuracy',cv=10)
grid_search = grid_search.fit(x_train,y_train)
```

```
grid_search.best_score_
```

```
grid_search.best_params_
```

```
from sklearn.svm import SVC

model = SVC(C=10,gamma='scale',kernel='rbf')
model.fit(x_train,y_train)
```

```
y_pred = model.predict(x_test)
y_pred
```

### **EVALUATION METRICS**

```
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

#Accuracy Score
print('ACCURACY SCORE : ',accuracy_score(y_pred,y_test)*100)
```

```
#Confusion Matrix

print('CONFUSION MATRIX \n',confusion_matrix(y_pred,y_test))
```

```
#Classification Report

print('CLASSIFICATION REPORT \n',classification_report(y_pred,y_test))
```
